# Credit Fraud Detection

## Specify the Question

Fraud could be one of the main reason's banks suffer from customer churning.It affects their credibility and their loss of faith by the customers.

In simple terms,credit fraud is identity theft.This cases have increasing year to year. In the US in 2016, there was a 16% increase from 2015.

Banks have implemented their methods of fraud detection ie Rule Based Systems.However,this method has become very inconsistent since new innovative fraud patterns are invented while the system is not improving.
Credit Fraud could also be in the form of duplicate transactions.

## Business Objective

Detect a fraud case.

## Metric of Success
90% + Accurate model that shall detect fraud cases 

Differentiate between change in human behavioural patterns and fraud cases.

## Context

The datasets contains transactions made by credit cards in September 2013 by european cardholders.
This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.

It contains only numerical input variables which are the result of a PCA transformation. Unfortunately, due to confidentiality issues, we cannot provide the original features and more background information about the data. Features V1, V2, â€¦ V28 are the principal components obtained with PCA, the only features which have not been transformed with PCA are 'Time' and 'Amount'. Feature 'Time' contains the seconds elapsed between each transaction and the first transaction in the dataset. The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-senstive learning. Feature 'Class' is the response variable and it takes value 1 in case of fraud and 0 otherwise.

# Modelling

## K-means

```python
#make a copy of the original dataframe
df1 = df.copy()
#drop the response variable
X = df1.drop(["Class"],axis=1)
y = df1["Class"]
#we have successfully dropped the column
```

```python
#put the
#data = df1.values[:, 0:30]
#data.values
```

```python
#implemention of k-means
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
#predict the model 
y_kmean = kmeans.predict(X)
#dff = pd.DataFrame(y_kmeans,columns=df.columns)
```

```python
#check the performance
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
print("***********CLASSIFICATION REPORT*****************")
print(classification_report(y,y_kmean))
print("***********F1_SCORE*******************")
print(f1_score(y,y_kmean))
```

## K means with scaled dataframe

```python
#split the dataset
#X = scaled_df.drop(["Class"],axis=1)
#y = scaled_df["Class"]
#drop the response variable
#scaled_df.drop(["Class"],axis=1)
X = scaled_df.iloc[:,:-1].values
y = scaled_df["Class"]
X.reshape(1,-1)
#scaled_df = scaled_df.drop(["Class"],axis=1,inplace=True)
#scaled_df
```

```python
X[0].shape
```

```python
X[1]
```

```python
kmeans = KMeans(n_clusters=2,random_state=11)

kmeans.fit(X)
```

```python
#check centres of clusters
kmeans.cluster_centers_
```

```python
#report back the label it believe to be true for the cluster
kmeans.labels_
```

```python
#predict
y_kmeans = kmeans.predict(X)
y_kmeans = pd.DataFrame(y_kmeans)
print(y_kmeans)
#y_kmeans.count()
display(y_kmeans[y_kmeans[0] == 1].count())

display(y_kmeans[y_kmeans[0] == 0].count())
#convert y_kmeans into arrray
ykmeans=y_kmeans.values.reshape(1,-1)
ykmeans
```

```python
#visualization
centers = kmeans.cluster_centers_

# Plot the clustered data
fig, ax = plt.subplots(figsize=(20, 10))
plt.scatter(X[kmeans.labels_ == 0, 0], X[kmeans.labels_ == 0, 1],s=50,cmap='OrRd',
            c='red', label='cluster 1')
plt.scatter(X[kmeans.labels_ == 1, 0], X[kmeans.labels_ == 1, 1],s=50,cmap='OrRd',
            c='orange', label='cluster 2')
plt.scatter(centers[:, 0], centers[:, 1], marker='*', s=300,alpha = 0.5,
            c='b', label='centers')
plt.legend();
#plt.xlim([-2, 2])
#plt.ylim([-2, 2])
#plt.xlabel('Eruption time in mins')
#plt.ylabel('Waiting time to next eruption')
#plt.title('Visualization of clustered data', fontweight='bold')
#ax.set_aspect('equal');
```

```python
#check the performance
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
print("***********CLASSIFICATION REPORT*****************")
print(classification_report(y,y_kmeans))
print("***********F1_SCORE*******************")
print(f1_score(y,y_kmeans))
```
